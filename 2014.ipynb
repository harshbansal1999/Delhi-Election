```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('tabula-2015.csv',header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns=['Pooling_Station_Number','Aditya','Rakesh_Kumar','Rohit_Tyagi','Sharmistha_Mukherjee','Saurabh_Bharadwaj',
            'Ashok_Kumar','Ashok_Sharma','Sanjay_Khatri','Total_Of_Valid_Votes','No._Of_Rejected_Votes','Total_Votes_For_NOTA','No._Of_Tendered_Votes']
```

```python
df.head()
```

```python
df.tail()
```

```python
total_votes=df.iloc[152]
```

```python
total_votes
```

```python
df.drop(152,inplace=True)
```

```python
df.tail()
```

```python
votes_pooling=pd.DataFrame(df.loc[:,['Pooling_Station_Number','Total_Of_Valid_Votes']])
```

```python
votes_pooling.head()
```

```python
stats=df.describe()
```

```python
stats
```

```python
stats.columns
```

```python
stats.drop(['Total_Of_Valid_Votes','No._Of_Rejected_Votes','Total_Votes_For_NOTA','No._Of_Tendered_Votes'],axis=1,inplace=True)
```

```python
stats
```

```python
candidate=['Aditya','Rakesh_Kumar','Rohit_Tyagi','Sharmistha_Mukherjee','Saurabh_Bharadwaj',
            'Ashok_Kumar','Ashok_Sharma','Sanjay_Khatri']
```

```python
candidate[1]
```

```python
votes=pd.DataFrame(total_votes,index=candidate)
votes.columns=['Total_Votes']
votes['Average_Votes_Per_Station']=stats.loc['mean',:]
votes['Maximum_Votes_From_A_Station']=stats.loc['max',:]
votes['Minimum_Votes_From_A_Station']=stats.loc['min',:]
```

```python
votes.head()
```

```python
max_list=[]
for i in range(len(candidate)):
    max_list.append(df[candidate[i]].idxmax())
```

```python
min_list=[]
for i in range(len(candidate)):
    min_list.append(df[candidate[i]].idxmin())
```

```python
votes['Maximum_Votes_Pooling_Station']=max_list
votes['Minimum_Votes_Pooling_Station']=min_list
```

```python
votes.head()
```

```python
plt.figure(figsize=(16,8))
plt.plot(votes_pooling['Pooling_Station_Number'],votes_pooling['Total_Of_Valid_Votes'])
plt.xticks(np.arange(1,164,10))
plt.xlabel('Pooling Station')
plt.ylabel('Total Votes')
plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(votes.index,votes['Total_Votes'])
plt.xlabel('Candidates')
plt.ylabel('Total Votes')
plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(20,8))
plt.scatter(votes.index,votes['Average_Votes_Per_Station'])
plt.xlabel('Candidates')
plt.ylabel('Average Votes Per Station')
plt.show()
plt.tight_layout()
```
